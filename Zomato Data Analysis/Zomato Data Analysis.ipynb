```python
country=pd.read_excel("./data/Country-Code.xlsx")
```

```python
country
```

```python
df=pd.merge(df,country,on="Country Code",how="outer")
```

```python
df.head()
```

```python
Names= df.Country.value_counts().index
Values=df.Country.value_counts().values
explode = [0.1, 0, 0, 0, 0]
# define Seaborn color palette to use
palette_color = sns.color_palette('pastel')
plt.figure(figsize=(10,10))
  
# plotting data on chart
plt.pie(Values[:5], labels=Names[:5], colors=palette_color,
        explode=explode, autopct='%.0f%%')
  
# displaying chart

plt.show()
```
